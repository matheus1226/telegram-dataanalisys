```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
file = pd.read_csv("../data/raw/Avaliação_Mensal_Funcionários.csv", decimal = ',')
file
```

```python
file.info()
```

```python
file.Setor = file.Setor.replace({"Engenheiro de Software":"Engenharia de Software"})
```

```python
setores = file.groupby("Setor").mean(numeric_only=True)["NPS interno"].index
nps_medio_setor = file.groupby("Setor").mean(numeric_only=True)["NPS interno"].values

plt.bar(setores,nps_medio_setor)
plt.ylabel("Nps interno mensal médio")
plt.xticks(rotation = 30)
plt.yticks(np.array(range(0,11,1)));
```

```python
tipo_contratacao = file.groupby("Tipo de Contratação").mean(numeric_only=True)["NPS interno"].index
nps_medio_por_tipo_contratacao = file.groupby("Tipo de Contratação").mean(numeric_only=True)["NPS interno"].values

plt.bar(tipo_contratacao,nps_medio_por_tipo_contratacao)
plt.ylabel("Nps interno por contratação mensal médio")
plt.xticks(rotation = 30)
plt.yticks(np.array(range(0,11,1)))
plt.title("Média de NPS interno Mensal por Tipo de Contratação");
```
